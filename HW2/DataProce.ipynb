```python
import pandas as pd
df = pd.read_csv('adult.data',header = None, sep=',',na_values=' ?')
print(df.head())
original_dataset=df
```

```python
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income']
df
```

```python
labels=["{0}-{1}".format(i,i+9)for i in range(0,100,20)]
labels
```

```python
df.loc[df.income ==" >50K", "encoding_income"]= int(1)
df.loc[df.income ==" <=50K", "encoding_income"]=int(0)
df["age"]=pd.cut(df.age,range(0,101,20), right=False,labels=labels)
```

```python
df
```

```python
import pandas_profiling
#df.profile_report()
```

```python
df_test=df[['age','education','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','encoding_income']]
df_test
```

```python
#對資料做One-hot encoding
df_test = pd.get_dummies(df_test,columns = ['age','education','marital-status','occupation','relationship','race','sex'])
df_test
```

```python
#將資料做Random
df_test=df_test.sample(frac=1)
```

```python
# X=df_test[['age','fnlwgt','education','capital-gain','capital-loss','hours-per-week']]
#設定欄位
X=df_test.drop('encoding_income', axis=1)
y=df_test['encoding_income']
```

```python
#切分資料
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3 ,random_state=1009)
```

```python
#訓練
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy',max_depth=7)   # criterion可設定為entropy, 決策深度設為 4
tree.fit(X_train, y_train)
```

```python
#訓練集正確率
tree.score(X_train,y_train)
```

```python
result_test = tree.predict(X_test.iloc[:,:])
result_train= tree.predict(X_train.iloc[:,:])
```

```python
#批次訓練
train_list=[]
test_list=[]
for i in range(1,21):
    #將資料做Random以及切分資料
    df_test=df_test.sample(frac=1)
    X=df_test.drop('encoding_income', axis=1)
    y=df_test['encoding_income']
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3 ,random_state=1009)
    tree = DecisionTreeClassifier(criterion='entropy',max_depth=i)   # criterion可設定為entropy, 決策深度設為 4
    tree.fit(X_train, y_train)
    score_test=tree.score(X_test, y_test)
    score_train=tree.score(X_train, y_train)
    test_list.append(score_test)
    train_list.append(score_train)

result={
    "train":train_list,
    "test":test_list
}
df_result=pd.DataFrame(result)
```

```python
import matplotlib
matplotlib.use('TKAgg') 
%matplotlib inline
import matplotlib.pyplot as plt


plt.plot(train_list,color=(255/255,100/255,100/255))
plt.plot(test_list,color=(100/255,100/255,255/255))

plt.ylabel("Accuracy")
plt.xlabel("Depth_of_tree")
plt.title("Adult_dataset_accuracy")
plt.xlim([0,21])
plt.grid(True)

plt.show()
```

```python
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., 1.02*height, '%f'%float(height),
                ha='center', va='bottom')
```
